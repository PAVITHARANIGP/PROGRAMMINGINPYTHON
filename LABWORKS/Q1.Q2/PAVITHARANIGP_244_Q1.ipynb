```python
import pandas as pd

us_cereal_df = pd.read_csv("UScereal.csv")

cereals_from_manufacturer_G = us_cereal_df[us_cereal_df['mfr'] == 'G']

highest_sugar_cereal = cereals_from_manufacturer_G[cereals_from_manufacturer_G['sugars'] == cereals_from_manufacturer_G['sugars'].max()]

name_of_highest_sugar_cereal = highest_sugar_cereal['Name'].values[0]  

print(f"The cereal with the highest sugar content from Manufacturer G is: {name_of_highest_sugar_cereal}")
```
